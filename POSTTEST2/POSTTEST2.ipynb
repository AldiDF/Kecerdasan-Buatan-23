### [5] Visualisasi Data BarPlot ###

```python
eventType_sum = df.groupby('event_type')['economic_impact_million_usd'].sum()

bars = plt.bar(eventType_sum.index, eventType_sum.values)

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval,                            
        f"{yval:,.0f}",
        ha="center", va="bottom", fontsize=8, rotation=0
    )

plt.title("Jumlah Kerugian Ekonomi Akibat Peristiwa Alam Dalam 5 Tahun Terakhir")
plt.xlabel("Jenis Peristiwa Alam") 
plt.ylabel("Dampak Kerugian Ekonomi (dalam Jutaan USD)") 
plt.xticks(rotation=90, ha="right")
plt.show()
```

### [6] Visualisasi Data HeatMap ###

```python
columns = ["severity", "duration_days", "affected_population", "economic_impact_million_usd", "total_casualties"]

cor_matrix = df[columns].corr()

plt.figure(figsize=(8,6))
sns.heatmap(cor_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi: Dampak Peristiwa Alam", fontsize=14, pad=12)
plt.show()
```

### [7] Kesimpulan Visualisasi Data ###

Berdasarkan visualisasi data dengan BarPlot, 'Heatwave' atau gelombang panas memiliki jumlah kerugian tertinggi dalam 5 tahun terakhir dengan jumlah kerugian 2157 juta USD, dan 'Volcano Eruption' atau erupsi gunung berapi memiliki jumlah kerugian terendah dalam 5 tahun terakhir dengan jumlah kerugian 26 juta USD.

Berdasarkan visualisasi data dengan HeatMap didapatkan hasil analisis sebagai berikut:

- Korelasi Rendah (0.1 - 0.9)
    - duration_days - affected_population = 0.01,
        Artinya durasi waktu peristiwa alam tidak selalu membuat lebih banyak orang terdampak. Bisa jadi ada peristiwa alam dengan tingkat kerusakan lebih parah namun durasi waktunya singkat atau ada yang kerusakannya lebih ringan namun durasinya lebih lama.
    - total_casualties - affected_population = 0.9, 
        Artinya jumlah orang terdampak tidak selalu jumlah korbannya juga tinggi.

- Korelasi Sedang (0.10 - 0.22)
