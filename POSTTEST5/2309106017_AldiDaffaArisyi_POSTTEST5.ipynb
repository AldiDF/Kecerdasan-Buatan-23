### [5] Visualisasi Data BarPlot ###

```python
eventType_sum = df.groupby('event_type')['economic_impact_million_usd'].sum()

bars = plt.bar(eventType_sum.index, eventType_sum.values)

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval,                            
        f"{yval:,.0f}",
        ha="center", va="bottom", fontsize=8, rotation=0
    )

plt.title("Jumlah Kerugian Ekonomi Akibat Peristiwa Alam Dalam 5 Tahun Terakhir")
plt.xlabel("Jenis Peristiwa Alam") 
plt.ylabel("Dampak Kerugian Ekonomi (dalam Jutaan USD)") 
plt.xticks(rotation=90, ha="right")
plt.show()
```

### [6] Visualisasi Data HeatMap ###

### [2] Scaling (Normalisasi/Standarisasi) ###

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#Numerical Columns
num_col_ = [col for col in df.columns if
df[col].dtypes != 'O' and col != 'Id']
x1 = df[num_col_]

x1 = scaler.fit_transform(x1)
x1 = pd.DataFrame(x1, columns=num_col_)
x1.head()
```

Melakukan Standarisasi untuk mengubah skala data

### [3] Encoding Kolom Kategorikal ###

```python
df_encoded = pd.get_dummies(df, columns=["event_type", "country"], drop_first=False)

df_encoded = df_encoded.drop(columns=num_col_)

df_final = pd.concat([x1, df_encoded], axis=1)

df_final.head()
```

Melakukan encoding untuk mengubah semua kolom kategorikal menjadi nilai boolean

### [4] Feature Engineering ###

```python
df_final['total_casualties'] = df_final['deaths'] + df_final['injuries']

df_final[['deaths', 'injuries', 'total_casualties']].head()
```

Membuat fitur baru df['total_casualties'] = df['deaths'] + df['injuries']

```python
df_final['impact_per_capita'] = df_final['economic_impact_million_usd'] / df_final['affected_population'].replace(0, 1)

df_final[['economic_impact_million_usd', 'affected_population', 'impact_per_capita']].head()
```

Membuat fitur baru df['impact_per_capita'] = df['economic_impact_million_usd'] / df['affected_population']

### Penerapan Deep Naural Network Pada Model

```python
import tensorflow as tf
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if(logs.get('r2_metric') > 0.84 and logs.get('val_r2_metric') > 0.84):
            self.model.stop_training = True


def r2_metric(y_true, y_pred):
    ss_res = tf.reduce_sum(tf.square(y_true - y_pred))
    ss_tot = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
    return 1 - ss_res / (ss_tot + tf.keras.backend.epsilon())

model_lr = tf.keras.Sequential([
    tf.keras.layers.Dense(units=256, activation='relu',),
    tf.keras.layers.Dense(units=128, activation='relu',),
    tf.keras.layers.Dense(units=64, activation='relu',),
    tf.keras.layers.Dense(units=32, activation='relu',),
    tf.keras.layers.Dense(units=1, activation='linear',)
])

model_lr.compile(optimizer='adam', loss='mean_squared_error', metrics=[r2_metric])

model_lr.fit(
    X_train, 
    y_train, 
    epochs=800, 
    validation_data=(X_test, y_test),
    callbacks=[MyCallback()]
)

y_pred_model_lr = model_lr.predict(X_test)

plt.scatter(y_test, y_pred_model_lr, color='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='blue', linewidth=3)
plt.xlabel("Nilai Aktual")
plt.ylabel("Nilai Prediksi")
plt.title("Aktual vs Prediksi Metode Linear Regression")
plt.show()

print("MAE:", mean_absolute_error(y_test, y_pred_model_lr))
print("MSE:", mean_squared_error(y_test, y_pred_model_lr))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_model_lr)))
print("R:", r2_score(y_test, y_pred_model_lr))
```

Berdasarkan hasil training selama 681 epoch (iterasi) dari 800 epoch didapatkan bahwa:
- loss: 0.0827, fungsi loss yang digunakan adalah MSE yang berarti rata-rata kesalahan kuadrat model terhadap data pelatihan cukup kecil (0.0827). Model belajar dengan baik dan tidak mengalami kesulitan besar memahami pola data training
- val_loss: 0.1366, lebih tinggi dari loss menandakan bahwa model agak sedikit lebih buruk pada data baru (yang tidak pernah dilihat), tetapi masih dalam batas wajar
- r2_metric: 0.9112, berarti model dapat menjelaskan 91,12% variasi data training
- val_r2_metric: 0.8453, berarti model mampu menjelaskan 84,53% variasi pada data validasi (data baru)

Berdasarkan gambar plot, sebagian besar titik hitam berada dekat dengan garis biru, yang berarti model cukup akurat dalam memprediksi

Berdasarkan hasil akhir (evaluasi) secara keseluruhan setelah model di training didapatkan bahwa:
